```python
import tkinter
from tkinter import *
from PIL import Image, ImageTk

root = Tk()
root.geometry("1000x1000")
root.configure(bg = "bisque1")
root.title("Menu")

welcome = tkinter.Label(root, text = "Welcome to Seaside Bistro!", font = "Vivaldi 30", width = 50, bg = "bisque1")
welcome.place(x = 790, y = 660, anchor = S)

path = Image.open("C:\\Users\\35987\\My Python Stuff\\Tkinter\\restaurant_image.jpg").resize((1600,600), Image.ANTIALIAS)
menu = ImageTk.PhotoImage(path)


label = tkinter.Label(image = menu)
label.image = menu
label.pack()


def open_breakfast():
    window = Toplevel(root)
    window.title("Pancakes")
    window.geometry("2200x2200")
    window.resizable(True, True)
    window.configure(bg='bisque1')
    
    breakfast_img = tkinter.Label(window, image = pancakes)
    breakfast_img.image = pancakes
    breakfast_img.pack(side = LEFT)
    
    
    ingredients = "Ingredients:\n100 g plain flour\n2 large eggs\n300 ml milk\n1 tbsp sunflower or vegetable oil\n"
    
    steps = """\nPreparation:
            \nSTEP 1\nPut 100 g plain flour, 2 large eggs, 300ml milk, 1 tbsp sunflower or vegetable oil \nand a pinch of salt into a bowl or large jug, then whisk to a smooth batter.\n
            \nSTEP 2\nSet aside for 30 mins to rest if you have time, or start cooking straight away.\n
            \nSTEP 3\nSet a medium frying pan or crêpe pan over a medium heat and carefully wipe it \nwith some oiled kitchen paper.\n
            \nSTEP 4\nWhen hot, cook your pancakes for 1 min on each side until golden, \nkeeping them warm in a low oven as you go.\n
            \nSTEP 5\nServe with lemon wedges and caster sugar, or your favourite filling."""
    
    
    prep = Canvas(window, width = 1500, height = 1600, bg = "bisque1")
    prep.create_text(355, 310, fill = "black", font = ("Elephant 12"), text = ingredients + steps, justify = "left")
    
    exit = Button(window, text = "Exit", width = 20, font = "Elephant 12", padx = 5, pady = 5, bg = "blanchedalmond", borderwidth = 5, command = window.destroy)
    exit.pack(side = BOTTOM, anchor = "center", padx = 5, pady = 5)
    
    lunch = Button(window, text = "Mushroom Ramen", width = 20, font = "Elephant 12", padx = 5, pady = 5, bg = "blanchedalmond", borderwidth = 5, command = open_lunch)
    lunch.pack(side = BOTTOM, anchor = "center", padx = 5, pady = 5)
    
    dinner = Button(window, text = "Sweet and sour chicken", width = 20, font = "Elephant 12", padx = 5, pady = 5, bg = "blanchedalmond", borderwidth = 5, command = open_dinner)
    dinner.pack(side = BOTTOM, anchor = "center", padx = 5, pady = 5)
    
    prep.pack()
  

                     
def open_lunch():
    window2 = Toplevel(root)
    window2.title("Mushroom Ramen")
    window2.geometry("2200x2200")
    window2.resizable(True, True)
    window2.configure(bg='bisque1')
    
    lunch_img = tkinter.Label(window2, image = ramen)
    lunch_img.image = ramen
    lunch_img.pack(side = LEFT)
    
    ingredients2 = """Ingredients:\n1/2 Tbsp cooking oil\n100 g baby bella mushrooms\n355 ml cups vegetable broth\n1 handful fresh spinach\n1 package ramen noodles\n120 ml coconut milk (canned)\n"""
    
    steps2 = """\nPreparation:
        \nSTEP 1\nSlice the mushrooms. Add them to a small sauce pot with the cooking oil and sauté \nover medium heat until the mushrooms are soft, dark, and all the moisture in the bottom of \nthe pot has evaporated.
            \nSTEP 2\nAdd the vegetable broth, turn the heat up to medium-high, and bring the broth up to a boil.\nOnce boiling, add the ramen noodles (without the seasoning packet) to the broth.\nCook the noodles in the boiling broth for about 3 minutes, or until tender.
            \nSTEP 3\nTurn the heat off, add a heaping handful of fresh spinach, and stir until the spinach is wilted. 
Pour the coconut milk into the pot and stir to combine.
            \nSTEP 4\nServe the Vegan Creamy Coconut Ramen as is, or with garnishes like green onion or chili garlic sauce."""
    
    
    prep2 = Canvas(window2, width = 1500, height = 1600, bg = "bisque1")
    prep2.create_text(435, 300, fill = "black", font = ("Elephant 12"), text = ingredients2 + steps2, justify = "left")
    
    exit = Button(window2, text = "Exit", width = 20, font = "Elephant 12", padx = 5, pady = 5, bg = "blanchedalmond", borderwidth = 5, command = window2.destroy)
    exit.pack(side = BOTTOM, anchor = "center", padx = 5, pady = 5)
    
    breakfast = Button(window2, text = "Pancakes", width = 20, font = "Elephant 12", padx = 5, pady = 5, bg = "blanchedalmond", borderwidth = 5, command = open_breakfast)
    breakfast.pack(side = BOTTOM, anchor = "center", padx = 10, pady = 10)
    
    dinner = Button(window2, text = "Sweet and sour chicken", width = 20, font = "Elephant 12", padx = 5, pady = 5, bg = "blanchedalmond", borderwidth = 5, command = open_dinner)
    dinner.pack(side = BOTTOM, anchor = "center", padx = 5, pady = 5)
    
    prep2.pack()
    

def open_dinner():
    window3 = Toplevel(root)
    window3.title("Sweet and sour chicken")
    window3.geometry("2200x2200")
    window3.resizable(True, True)
    window3.configure(bg='bisque1')
    
    dinner_img = tkinter.Label(window3, image = chicken)
    dinner_img.image = chicken
    dinner_img.pack(side = LEFT)
    
    
    ingredients3 = """Ingredients:\n5 tbsp vegetable oil\n2 eggs lightly beaten\n3 tbsp cornflour (cornstarch)\n90g plain flour\n1/2tsp salt\n1/2tsp pepper\n1/2tsp garlic salt\n2 tsp paprika\n3 chicken breast fillets\n\nSauce:\n1 tbsp vegetable oil\n1 large onion\n1 red pepper\n1 green pepper\n2 cloves of garlic\n1 tsp minced ginger\n150ml tomato ketchup\n12 tbsp malt vinegar\n6 tbsp dark brown muscovado sugar\n475 g canned of pineapple chunks in juice"""
    
    steps3 = """
            \nPreparation:
            \nSTEP 1\nHeat the oil in a wok or large frying pan until very hot.
            \nSTEP 2\nWhilst the oil is heating, place the egg in one shallow bowl and the cornflour in another shallow bowl.\nAdd the flour, salt, pepper, garlic salt and paprika to another shallow bowl and mix together.
            \nSTEP 3\nDredge the chicken in the cornflour, then dip in the egg (make sure all of the chicken is covered in egg wash), and finally dredge it in the seasoned flour.\nAdd to the wok and cook on a high heat for 6-7 minutes, turning two or three times during cooking, until well browned. 
            \nSTEP 4\nAdd the oil to a large frying pan/skillet, heat it up on a medium-to-high heat and add in the onions. Cook for 3 minutes, stirring every now and then, until the onions start to go translucent. Add the peppers and cook for a further minute. Add in the garlic and ginger and cook for another minute. Add the tomato ketchup, vinegar, dark brown sugar, and the tinned pineapple
            \nSTEP 5\nTransfer the cooked crispy chicken to the pan with the sauce and toss to coat. Serve with boiled or fried rice and a sprinkling of sesame seeds."""
    
    
    prep3 = Canvas(window3, width = 1500, height = 1600, bg = "bisque1", scrollregion = (0,0,1350,1050))
    prep3.create_text(1450, 450, fill = "black", font = ("Elephant 12"), text = ingredients3 + steps3, justify = "left")
    
    exit = Button(window3, text = "Exit", width = 20, font = "Elephant 12", padx = 5, pady = 5, bg = "blanchedalmond", borderwidth = 5, command = window3.destroy)
    exit.pack(side = BOTTOM, anchor = "center", padx = 5, pady = 5)
    
    breakfast = Button(window3, text = "Pancakes", width = 20, font = "Elephant 12", padx = 5, pady = 5, bg = "blanchedalmond", borderwidth = 5, command = open_breakfast)
    breakfast.pack(side = BOTTOM, anchor = "center", padx = 10, pady = 10)
    
    lunch = Button(window3, text = "Mushroom Ramen", width = 20, font = "Elephant 12", padx = 5, pady = 5, bg = "blanchedalmond", borderwidth = 5, command = open_lunch)
    lunch.pack(side = BOTTOM, anchor = "center", padx = 5, pady = 5)
    
    
    vertical_scrollbar = tkinter.Scrollbar(window3, orient = VERTICAL, command = prep3.yview)
    vertical_scrollbar.pack(side = RIGHT)
    
    horizontal_scrollbar = tkinter.Scrollbar(window3, orient = HORIZONTAL, command = prep3.xview)
    horizontal_scrollbar.pack(side = BOTTOM)
    
    prep3["yscrollcommand"] = vertical_scrollbar.set
    prep3["xscrollcommand"] = horizontal_scrollbar.set
    
    prep3.pack()
    
    
    
path2 = Image.open("C:\\Users\\35987\\My Python Stuff\\Tkinter\\pancakes.jpg").resize((700,800), Image.ANTIALIAS)
pancakes = ImageTk.PhotoImage(path2)  

path3 = Image.open("C:\\Users\\35987\\My Python Stuff\\Tkinter\\mushroom_ramen.jpg").resize((700,800), Image.ANTIALIAS)
ramen = ImageTk.PhotoImage(path3)

path4 = Image.open("C:\\Users\\35987\\My Python Stuff\\Tkinter\\sweet_and_sour_chicken.jpg").resize((700,800), Image.ANTIALIAS)
chicken = ImageTk.PhotoImage(path4) 


breakfast = Button(root, text = "Pancakes", font = "Elephant 12", padx = 5, pady = 5, width = 15, bg = "blanchedalmond", borderwidth = 5, command = open_breakfast)
breakfast.place(x = 510 , y = 720, anchor = SE)

lunch = Button(root, text = "Ramen", font = "Elephant 12", padx = 5, pady = 5, width = 15, bg = "blanchedalmond", borderwidth = 5, command = open_lunch)
lunch.place(x = 780, y = 720, anchor = S)

dinner = Button(root, text = "Sweet and sour chicken", font = "Elephant 12", padx = 5, pady = 5, width = 20, bg = "blanchedalmond", borderwidth = 5, command = open_dinner)
dinner.place(x = 1040, y = 720, anchor = SW)

exit = Button(root, text = "Exit", font = "Elephant 12", padx = 5, pady = 5, width = 15, bg = "blanchedalmond", borderwidth = 5, command = root.destroy)
exit.place(x = 780, y = 780, anchor = S)

root.mainloop()
```
